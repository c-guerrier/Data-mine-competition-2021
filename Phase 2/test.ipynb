```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel('Diabete Cleaned data set2.xlsx', sheet_name = 0, index_col = 0)
```

```python
df.head()
```

```python
df.T
```

```python
X = df.iloc[:, 8].values.reshape(-1, 1)  # values converts it into a numpy array
Y = df.iloc[:, 10].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
model = linear_regressor.fit(X, Y)  # perform linear regression
Y_pred = linear_regressor.predict(X)  # make predictions
```

```python
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red', marker='o',
     markerfacecolor='blue', markersize=12)
plt.show()
```

```python
Y_pred
```

```python
model.score(X,Y)
```
